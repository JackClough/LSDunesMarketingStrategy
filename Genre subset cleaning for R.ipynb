```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.special import logit, expit
```

```python
df = pd.read_csv('genre_subset.csv')
df = df.drop(columns = 'Unnamed: 0')
```

```python
df['key'] = df['key'].astype('category')
df['mode'] = df['mode'].astype('category')
```

```python
df['energy'] = df['energy'].apply(lambda x: x/100)
```

```python
cols = df.columns.tolist()
```

```python
cols = ['Artist',
 'Song',
 'Genres',
 'Top_Genre',
 'danceability',
 'energy',
 'key',
 'loudness',
 'mode',
 'acousticness',
 'instrumentalness',
 'valence',
 'tempo',
 'duration_minutes',
 'Lyrics',
 'Total Words',
 'Total Unique Words',
 'Avg. Word Length',
 'Top 10 Words',
 'Lexical Diversity',
 'popularity']
```

```python
df = df[cols]
```

```python
df.hist(bins=30, figsize=(15, 10))
```

```python
df['energy'] = logit(df['energy']-.001)
```

```python
df.hist(bins=30, figsize=(15, 10))
```

```python
df.to_csv('songs_transformed.csv')
```
